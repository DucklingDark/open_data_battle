Добавим класс, в котором объединим признаки и целевой признак. Это немного упростит код и улучшит читаемость

```python
class ML_data:
    def __init__(self, X, y):
        self.X = X
        self.y = y
```

Добавим класс `Helper`<br>

```python
class Helper:
    def __init__(self):
        pass
    
    def print_random_rows(self, df, n):
        print('----- Данные -----')
        display(df.sample(n=n, random_state=SEED))
    
    def print_describe(self, df):
        print('----- Статистика о данных -----')
        display(df.describe().T)
        
    def print_info(self, df):
        print('----- Типы данных -----')
        df.info(memory_usage='deep')
        
    def print_nans(self, df):
        print('----- NaN значения -----')
        nans_info = pd.DataFrame()
        nans_info['Количество'] = df.isnull().sum()
        nans_info['Процент от общего количества'] = (nans_info['Количество'] / df.shape[0] * 100).round(2)
        nans_info = nans_info.sort_values(by='Количество', ascending=False)
        display(nans_info) 
    
    def print_unique_values(self, df, column):
        print('----- Уникальные значения столбца', column, '-----')
        print(df[column].unique())
    
    def print_mean_median(self, column):
        print('Медианное значение: {:.2f}'.format(column.median()))
        print('Среднее значение: {:.2f}'.format(column.mean()))
    
    def draw_boxplot(self, df, columns, 
                     title, xlabel='', 
                     figsize=(14, 6), ylim=None):
        fig = plt.figure(figsize=figsize)
        ax = sns.boxplot(data=df[columns], color='red')
        ax.set(title=title, xlabel=xlabel)
        if ylim != None:
            plt.ylim(ylim)
            
    def draw_countplot(self, df, x, title, figsize=(14,5),
                       xlabel='', ylabel='', xticks=None):
        plt.figure(figsize=figsize)
        ax = sns.countplot(data=df, x=x)
        ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
        if (xticks!=None):
            plt.xticks(range(len(xticks)), xticks)
    
    def draw_histogram(self, df, column, title, kde=True, 
                       figsize=(14, 6), bins=None, xlabel=''):
        plt.figure(figsize=figsize)
        ax = sns.distplot(df[column], bins=bins, kde=kde, hist=True, color='red')
        ax.set(title=title, xlabel=xlabel)
        
    def draw_relplot(self, df, x, y, title, 
                     xlabel='', ylabel=''):
        ax = sns.relplot(data=df, x=x, y=y)
        ax.set(xlabel=xlabel, ylabel=ylabel, title=title)
    
    def print_grouped_statistics(self, df, column, agg_columns):
        display(df.groupby(column)[agg_columns].agg(['min', 'max', 
                                                     'mean', 'median']).sort_values(by='median'))
    
    def elements_percent(self, title, first, second):
        print(title, '{:.0%}'.format(first.shape[0]/second.shape[0]))
    
    def split_train_test(self, X, y):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=SEED)
        train = ML_data(X_train, y_train)
        test = ML_data(X_test, y_test)
        return train, test
    
    def calculate_metrics(self, target, predictions):
        mae = mean_absolute_error(target, predictions)
        rmse = mean_squared_error(target, predictions, squared=False)
        return mae, rmse
    
    def fit_and_predict(self, model, train: ML_data, test: ML_data):
        start = time.time()
        model.fit(X=train.X, y=train.y)
        fit_time = time.time() - start
        start = time.time()
        predictions = model.predict(test.X)
        prediction_time = time.time() - start
        predicted = pd.Series(data=predictions, index=test.X.index)
        rmse = mean_squared_error(test.y, predictions, squared=False)
        return round(fit_time, 4), round(prediction_time, 4), predicted, round(rmse, 2)
    
helper = Helper()
```

### <font color='green'><b>Описание данных</b></font>

Опираясь на дополнительную информацию из файла с историей, опишем данные:
- `carat_weight` - вес бриллианта в каратах. Цена и редкость бриллианта могут расти экспоненциально вместе с его размером. Например, один камень в два карата может стоить в два раза больше, чем два камня в один карат (при тех же параметрах огранки, чистоты и цвета)
- `cut` - огранка камня. "Огранка Толстовского" - это описание вариантов геометрических пропорций и размеров бриллианта, которые позволяют математически максимизировать светопреломление камней. Таким образом, этот параметр является самым важнейшим, влияющим на качество и цену камня. Варианты качества огранки:
    - `poor` - плохое
    - `fair` - посредственное
    - `good` - хорошее
    - `very good` - очень хорошее
    - `ideal` - идеальное
    - `signature ideal` - безупречно идеальное
- `color` - цвет. Цвет бриллианта играет роль светофильтра, что влияет на преломление света и "сверкание" камня. Шкала цвета:
    - `D` - совершенно бесцветные. Крайне редкие камни, цвет в них не обнаруживается
    - `E` - бесцветные. Содержат мельчайшие следы цвета, которые обнаруживаются только специалистами-геммологами
    - `F` - бесцветные. Незначительные следы цвета, которые также обнаруживаются только специалистами-геммологами
    - `G-H` - почти бесцветные. Цвет обнаруживается при сравнении с бриллиантом высшего качества
    - `I-J` - почти бесцветные. Незначительный выраженный цвет, который обнаруживается глазом неспециалиста
    - `K-Z` - заметный цвет. Легко обнаруживается глазом неспециалиста
- `clarity` - чистота. Различные минеральные отложения и неоднородности ухудшают преломления света, а соответственно - снижают стоимость камня. По чистоте различают следующие типы:
    - `FL/IF` - flawless (без дефектов) и internally flawless (без внутренних дефектов)
    - `VVS1/VVS2` - very, very slight inclusions (крайне незначительные включения)
    - `VS1/VS2` - very slight inclusions (очень незначительные включения)
    - `SI1/SI2` - slight inclusions (незначительные включения)
    - `I1/I2/I3` - самые плохие варианты
- `polish и symmetry` - полировка и симметрия влияют на то, как камень преломляет и отражает свет, и оказывают некоторое воздействие на качество и цену камня, так как они учитываются при оценке огранки и цвета. Используются следующие значения
    - `ID` - ideal (идеальное). Используется в сертификатах `AGSL`
    - `EX` - excellent (прекрасное)
    - `VG` - very good (очень хорошее)
    - `G` - good (хорошее)
- `report` - тип сертификата. Наличие сертификата гарантирует его качество и потенциально увеличивает стоимость
    - `AGSL` - American Gemologidal Society Laboratury
    - `GIA` - Gemological Institute of America
- `price` - цена бриллианта

Из описания данных, можно сделать вывод, что каждый параметр в той или иной мере влияет на потенциальную стоимость камня

### Преобразование признаков

Перед началом обучения моделей преобразуем категориальные данные в числовые. Для этого, попробуем два варианта преобразования:
- one-hot encoding. Заменим один признак на несколько (n-1, где n - количество вариантов признака). Столбцы будут содержать значения 0 (нет), 1 (да). Используем также параметр `drop_first=True`, чтобы не попасть не создавать лишний признак и не попасть в "dummy ловушку"
- label encoding. Заменим категориальные значения признака на числовые так, что значения буду в диапазоне `0..n-1` (где n - количество вариантов признака). Не будем использовать готовую функцию библиотеки `sklearn`, а напишем свои функции, в которых явно укажем "градацию" признака (0 - самое низкое качество, n-1 - самое высокое качество)

#### One-hot кодирование

```python
X_one_hot = pd.get_dummies(X, drop_first=True)
X_one_hot
```

#### Label кодирование

```python
def get_cut(value):
    if value == 'Fair':
        return 0
    elif value == 'Good':
        return 1
    elif value == 'Very Good':
        return 2
    elif value == 'Ideal':
        return 3
    elif value == 'Signature-Ideal':
        return 4
```

При преобразовани цвета не будем объединять некоторые буквы вместе, как указано в документе, предполагая, что хоть буквы находятся в одном классе, но одна чуть лучше другой (`G-H`, предполагаем, что `G` чуть лучше `H`, графики это подтверждают)

```python
def get_color(value):
    if value == 'I':
        return 0
    elif value == 'H':
        return 1
    elif value == 'G':
        return 2
    elif value == 'F':
        return 3
    elif value == 'E':
        return 4
    elif value == 'D':
        return 5
```

С чистотой поступим также (предполагаем, что `VS1` лчше `VS2`, графики это также подтверждают)

```python
def get_clarity(value):
    if value == 'SI2':
        return 0
    if value == 'SI1':
        return 1
    elif value == 'VS2':
        return 2
    elif value == 'VS1':
        return 3
    elif value == 'VVS2':
        return 4
    elif value == 'VVS1':
        return 5
    elif value == 'IF':
        return 6
    elif value == 'FL':
        return 7
```

Полировку и симметрию объединим в одну функцию, так как варианты значений одинаковые

```python
def get_polish_or_symmetry(value):
    if value == 'G':
        return 0
    elif value == 'VG':
        return 1
    elif value == 'EX':
        return 2
    elif value == 'ID':
        return 3
```

```python
def get_report(value):
    if value == 'GIA':
        return 0
    elif value =='AGSL':
        return 1
```

Теперь преобразуем данные

```python
X_labeled = X.copy()
X_labeled['cut'] = X_labeled['cut'].apply(get_cut)
X_labeled['color'] = X_labeled['color'].apply(get_color)
X_labeled['clarity'] = X_labeled['clarity'].apply(get_clarity)
X_labeled['polish'] = X_labeled['polish'].apply(get_polish_or_symmetry)
X_labeled['symmetry'] = X_labeled['symmetry'].apply(get_polish_or_symmetry)
X_labeled['report'] = X_labeled['report'].apply(get_report)
X_labeled
```

#### Дерево решений

```python
def try_decision_tree(train: ML_data, test: ML_data):
    top_rmse = 1000000
    top_fit_time = 0
    top_pred_time = 0
    top_depth = 0

    for depth in range (1,20):
        model = DecisionTreeRegressor(max_depth=depth, random_state=SEED)
        fit_time, prediction_time, predicted, rmse = helper.fit_and_predict(model, 
                                                                            train=train, 
                                                                            test=test)
        if (rmse < top_rmse):
            top_rmse = rmse
            top_fit_time = fit_time
            top_pred_time = prediction_time
            top_depth = depth
         
    print('Лучшая метрика RMSE:', top_rmse)
    print('Глубина:', top_depth)
    print('Время обучения:', top_fit_time)
    print('Время предсказания:', top_pred_time)
```

#### Случайный лес

```python
def try_random_forest(train: ML_data, test: ML_data):
    top_rmse = 1000000
    top_fit_time = 0
    top_pred_time = 0
    top_depth = 0
    top_estimators = 0

    for depth in range (1,20):
        for est in range(100, 300, 25):
            model = RandomForestRegressor(max_depth=depth, n_estimators=est, random_state=SEED)
            fit_time, prediction_time, predicted, rmse = helper.fit_and_predict(model, 
                                                                                train=train, 
                                                                                test=test)
            if (rmse < top_rmse):
                top_rmse = rmse
                top_fit_time = fit_time
                top_pred_time = prediction_time
                top_depth = depth
                top_estimators = est

    print('Лучшая метрика RMSE:', top_rmse)
    print('Глубина:', top_depth)
    print('Оценщиков:', top_estimators)
    print('Время обучения:', top_fit_time)
    print('Время предсказания:', top_pred_time)
```
